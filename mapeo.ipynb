# Oscilador

$$y''+ \omega_{0}sin(y)=g(t)$$
pero
$$y'=u$$
$$\Rightarrow$$
$$u'=-\omega_{0}^2sin(y)+g(t)$$

```python
from pylab import *
%run "mis_int.ipynb"
%matplotlib inline
```

```python
def osc(y,t,w02=1.):
    #return array([y[1],-w02*sin(y[0])]
    return array([y[1],-w02*sin(y[0])+0.1*sin(t)])
```

```python
t,sol =rk4(osc,[1.,0.],[0,10])
zoom = 2
xlim([-zoom,zoom]),ylim([-zoom,zoom])
axes().set_aspect('equal'),grid(True)
plot(sol[:,0],sol[:,1],'.')
show()
```

```python
for i in range(120):
    y0,u0 = pi*(2*rand()-1),0 #4*(2*rand()-1)
    t,sol =rk4(osc,[y0,u0],[0,50.])
    plot(sol[:,0],sol[:,1])
zoom = pi*3
xlim([-zoom,zoom]),ylim([-zoom,zoom])
axes().set_aspect('equal'),grid(True)
show()
```
