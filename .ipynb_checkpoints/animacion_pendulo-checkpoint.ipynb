```python
from pylab import *
from scipy.integrate import *
%matplotlib inline
```

```python
from JSAnimation import IPython_display
from matplotlib import animation
```

```python
def osc(y_v,t, ommega2=1,mu=0):
    dy = y_v[1]
    dv = -ommega2*sin(y_v[0])-mu*y_v[1]
    return array([dy,dv])
```

```python
t = linspace(0,100,500)
inicio = [0.,1.]
sol = odeint(osc, inicio, t, args=(1,0.1))
xlim([-1,1]),ylim([-1,1])
plot(sol[:,0], sol[:,1])
```

```python
figura = figure()
ejes = axes(xlim=(-1,1),ylim=(-1,1))

trayectoria, = ejes.plot([],[],lw=2)
punto, = ejes.plot([],[],'o')

def init():
    trayectoria.set_data([],[])
    punto.set_data([],[])
    return trayectoria, punto,

y,v = sol[:,0],sol[:,1]

def el_movimiento(i):
    trayectoria.set_data(y[:i+1],v[:i+1])
    punto.set_data(y[i],v[i])
    return trayectoria, punto,

animation.FuncAnimation(figura,el_movimiento,init_func=init,frames=100,interval=40)
```
