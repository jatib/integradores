## Problema 2

Haga una función que obtenga la derivada numérica de una función utilizando...

```python
def deriv(f,x,h=1):
    f_prim = (f(x+h)-f(x))/h
    return f_prim
```

```python
def func(x):
    return x**3-5*x+2

def deriv_analitica(x):
    return 3*(x**2)-5
```

```python
print deriv(func,10)
print deriv_analitica(10)

var_h = [1,.5,.1]
t = arange(0,10,0.1)

for n in var_h:
    plot(deriv(func,t,n),label='Derivada numerica para h=%s'%(n))

plot(deriv(func,t),label='Derivada numerica')

plot(deriv_analitica(t),label='Derivada analitica')
plot(func(t),label='Funcion')
grid(True)
legend()
show()
```

## Problema 3

Programe el método de Newton..

```python
def Newton(f,x_n,tol=0.5):
    raiz = 0
    x_nmas1 = x_n-(f(x_n)/deriv(f,x_n))
    if x_nmas1 <= tol:
        raiz = x_nmas1 
    return raiz
```

```python
def funcN(x):
    return sin(3*x)+0.5*(cos(2*x)*cos(2*x))
```

```python
time = arange(0,pi,0.1)

for t in time:
    Newton(funcN,t)
    #print t
```

## Problema 5

Haga un programa que calcule la inversa de una matriz 

```python
def matrizinv(matriz): 
    #donde se introduce la matriz como arreglo de 4 entradas entrada 0,1
    #son el renglon 1, entradas 2,3 son el renglon 2
    det = matriz[0]*matriz[3]-matriz[1]*matriz[2]
    if det != 0:
        inversa = [(det*matriz[3]),((-1)*det*matriz[1]),((-1)*det*matriz[2]),(det*matriz[0])]
    else:
        inversa = "No tiene matriz inversa"
    return inversa
```

#### Ingresa tu matriz de 2 x 2 como un vector ie (9,8,7,6)

```python
mat = [1,0,1,0]
matrizinv(mat)
```

## Problema 6

Haga una función que se llame suma numeros y que realice la suma de todos los dígitos de un número introducido

### No importa si tiene punto decimal o es negativo, esta cosa furula (blackbox XD)

```python
def sumar_numeros(x):
    
    ptos_decimales = 0
    decimales = 0
    suma = 0
    
    if x<0:
        x = x*(-1)
        longitud = str(x)
    else:
        longitud = str(x)
        
    for i in longitud:
        if i == '.':
            ptos_decimales+=1
        if ptos_decimales != 0:
            decimales += 1
    
    if decimales == 0:
        potencia = len(longitud)
        while potencia != -1:
            numero = x / 10**potencia
            suma = suma + numero
            x = x - numero*(10**potencia)
            potencia = potencia - 1
    else:
        potencia = (len(longitud))-1
        x = x*(10**(decimales-1))
        x = int(x)
        while potencia != -1:
            numero = x / 10**potencia
            suma = suma + numero
            x = x - numero*(10**potencia)
            potencia = potencia - 1
    return suma
```
